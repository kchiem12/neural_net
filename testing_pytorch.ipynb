```python
training_dataset = datasets.FashionMNIST(root="data", train=True, download=True, transform=ToTensor())  # To get the training set

labels_map = {     # This is to better label the images from the dataset
    0: "T-Shirt",
    1: "Trouser",
    2: "Pullover",
    3: "Dress",
    4: "Coat",
    5: "Sandal",
    6: "Shirt",
    7: "Sneaker",
    8: "Bag",
    9: "Ankle Boot",
}

figure = plt.figure(figsize=(8,8))
cols, rows = 3, 3
for i in range(1, rows*cols + 1):
    sample_index = torch.randint(len(training_dataset), size=(1,)).item()   # Gets a random index to choose from the dataset
    img, label = training_dataset[sample_index]
    figure.add_subplot(rows, cols, i)
    plt.title(labels_map[label])   # The label is any number from 0-9
    plt.axis("off")
    plt.imshow(img.squeeze(), cmap="bone")   # img.squeeze() removes any 1-dimensional elements in image
plt.show() 
```

```python
class RailroadsDataset(Dataset):
    def __init__(self, annotation_file, img_dir, transform=None, target_transformation=None):
        self.img_labels =  pd.read_csv(annotation_file)
        self.img_dir = img_dir
        self.transform = transform
        self.target_transform = target_transformation
    
    def __len__(self):
        return len(self.img_labels)
    
    def __getitem__(self, index):
        img_path = os.path.join(self.img_dir, self.img_labels.iloc[index, 1])
        image = read_image(img_path)
        label = self.img_labels.iloc[index, 2]
        if self.transform:
            image = self.transform(image)
        if self.target_transform:
            label = self.target_transform(label)
        return image, label
```
